## Read and process Diamon rate.txt output file

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
path = "test_data/rate.txt"

# read the file in
df = pd.read_csv(path, sep='\t', index_col=None)

# now sort of column headings
cols = df.columns
df=df.drop(cols[-1], 1)
cols = cols[1:]
df.columns = cols

df
```

```python
# add cumulative time
df['time'] = df['Dt(sec)'].cumsum()
# plot counts over time
plt.step(df["time"], df["Det1"])
plt.xlabel("Time (s)")
plt.ylabel("Counts")
```

```python
df['d1_sum'] = df['Det1'].cumsum()
# plot counts over time
plt.step(df["time"], df["d1_sum"])
plt.xlabel("Time (s)")
plt.ylabel("Counts")
```
