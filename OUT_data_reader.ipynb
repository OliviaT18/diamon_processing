## Data Reader Function 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
OUT_path = "test_data/OUT_data.txt"
```

```python
def read_file_to_df(path):
    
    df = pd.read_csv(path, sep='\t')
    df.drop(['Unnamed: 17'], axis = 1, inplace = True)
    cols = df.columns
    df = df.drop(cols[0] , axis = 1)
    cols = cols[1:]
    df.columns = cols
    
    return(df)
```

```python
OUT_df = read_file_to_df(OUT_path)
OUT_df
```

## Plotting Function

```python
def plot_df(data_frame, x, y, kind, x_err = None, y_err = None):
    df_graph = data_frame.plot(x = x, y = y, kind = kind, xerr = x_err, yerr = y_err)
    
    return (df_graph)
```
