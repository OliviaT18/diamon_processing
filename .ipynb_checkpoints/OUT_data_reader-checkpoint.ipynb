## Data Reader Function 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
OUT_path = "test_data/OUT_data.txt"
```

```python
def read_file_to_df(path):
    
    df = pd.read_csv(path, sep='\t')
    df.drop(['Unnamed: 17'], axis = 1, inplace = True)
    cols = df.columns
    df = df.drop(cols[0] , axis = 1)
    cols = cols[1:]
    df.columns = cols
    df = df.replace('\%','',regex=True)
    for x in df.columns:
        if '%' in x:
            df[x] = df[x].astype(float)

    return(df)
```

```python
OUT_df = read_file_to_df(OUT_path)
OUT_df
```

## Uncertainty to Absolute Error Function


```python
def un_to_abs_err(data_frame, value, un, col_name):
    
    data_frame[col_name] = (data_frame[un])/100 * data_frame[value]
    
    return(data_frame)
```

```python
OUT_df_new = un_to_abs_err(OUT_df,'H*(10)r', 'Hr_un%', 'Hr_err')
OUT_df_new
```

## Plotting Function

```python
def plot_df(data_frame, x, y, kind, x_err = None, y_err = None):
    
    df_graph = data_frame.plot(x = x, y = y, kind = kind, xerr = x_err, yerr = y_err)
    
    if '*' in x:
        df_graph.set_xlabel(x + '  uSv/h') #add in if and else for other columns once units known
    else:
        df_graph.set_xlabel(x)
    if '*' in y:
        df_graph.set_ylabel(y + '  uSv/h')
    else:
        df_graph.set_ylabel(y)
    
    return (df_graph)
```

```python
plot_df(OUT_df_new, 't(s)', 'H*(10)r', 'scatter', x_err = None, y_err = 'Hr_err')
```

## Stacked Bar Plot Function

```python
def stack_bar_plot(data_frame, cols, xlabel, ylabel):
    stack_df = data_frame.filter(cols)
    ax = stack_df.plot(kind='bar', stacked=True, figsize = (12,8))
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    
    return(ax)
```

```python
col = ['Ther%', 'Epit%', 'Fast%']
stack_bar_plot(OUT_df, col, 'index', '%')
```
